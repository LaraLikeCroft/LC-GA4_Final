```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://dreamlightvalleywiki.com/Fishing'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
#soup.find('table')
```

```python
find_table = soup.find_all('table')[1]
#print(find_table)
```

```python
find_fish_table = soup.find('table', class_ = 'wikitable sortable')
##<table class="wikitable sortable">
#print(find_fish_table)
```

```python
table = soup.find_all('table')[1]
#print(table)
```

```python
fish = table.find_all('th')
```

```python
fish_headers = [headers.text.strip() for headers in fish]
print(fish_headers)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = fish_headers)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    indv_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = indv_row_data
```

```python
df
```

```python
#df.drop_duplicates()
df = df.drop(columns=['Image', 'Ripples', 'Energy', 'Notes'])
df
```

```python
df['CookingCategory'] = "Fish"
df = df[['Name', 'CookingCategory', 'Sell Price', 'Locations']]
df
```

```python
df.to_csv('Web_Scrape_DDV_Fandom_Fish(Trimmed).csv')
```
